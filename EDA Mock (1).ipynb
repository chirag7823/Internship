```python
import pandas as pd
df=pd.read_csv(r'/home/nineleaps/Downloads/05_credit_card_transactions.csv')
df.head()
```

```python
# revenue generated by each merchant
import matplotlib.pyplot as plt
df.groupby('merchant')['amount'].sum().plot(kind='barh')
plt.title('Merchant Revenue')
plt.xlabel('Revenue')
plt.ylabel('Merchant')
plt.show()
```

```python
df.head()
```

```python
df.groupby('category')['amount'].mean().plot(kind='bar')
plt.title('Average Revenue Generated by each category')
plt.xlabel('Category')
plt.ylabel('Average Revenue')
plt.show()
```

```python
df.head()
```

```python
import seaborn as sns
sns.boxplot(df['amount'])
plt.show()
```

```python
df=df.sort_values('txn_date')
df.groupby('txn_date')['amount'].sum().plot(kind='line')
plt.title('Revenue trend line')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

```python
df.groupby('merchant')['amount'].mean().sort_values(ascending=False).head(5).plot(kind='bar')
plt.title('Top 5 Merchants ')
plt.xlabel('Merchant')
plt.ylabel('Average Revenue Generated')
plt.show()
```

```python
df['card_id'].value_counts().sort_values(ascending=False).head(5).plot(kind='barh')
plt.title('Highest buying cards')
plt.xlabel('Purchases')
plt.ylabel('Card ID')
plt.show()

```
