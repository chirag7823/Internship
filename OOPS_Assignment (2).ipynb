```python
class Product:
    def __init__(self,name,units,price):
        self.name = name
        self.units = units
        self.price = price
    def total_sales(self):
        return self.units * self.price
```

```python
class SalesReport:
    
    def __init__(self):
        self.products = []
    
    def add_product(self, product):
        self.products.append(product)
    
    def total_revenue(self):
        revenue = 0
        for p in self.products:
            revenue += p.total_sales()
        return revenue
    
    def highest_product(self):
        highest = self.products[0]
        for p in self.products:
            if p.total_sales() > highest.total_sales():
                highest = p
        return highest
    
    def print_report(self):
        print("Daily Sales Report")
        print("-"*50)
        print(f"{'Product':<15}{'Units':<10}{'Price':<10}{'Total':<10}")
        print("-"*50)
        for p in self.sort_products():
            total = p.total_sales()
            print(f"{p.name:<15}{p.units:<10}{p.price:<10.2f}{total:<10.2f}")
        print("-"*50)
        print(f"Total Revenue: {self.total_revenue():.2f}")

        highest = self.highest_product()
        print(f"Highest Selling Product: {highest.name} (Rs{highest.total_sales():.2f})")

    def sort_products(self):
        return sorted(self.products, key=lambda p: p.total_sales(), reverse=True)

    def save_report(self):
        with open("sales_report.txt", "w") as file:
            file.write("Daily Sales Report\n")
            file.write("-"*50 + "\n")
            for p in self.sort_products():
                total = p.total_sales()
                file.write(f"{p.name:<15}{p.units:<10}{p.price:<10.2f}{total:<10.2f}\n")
            file.write("-"*50 + "\n")
            file.write(f"Total Revenue: {self.total_revenue():.2f}\n")
```

```python
report = SalesReport()
while True:
    name = input("Enter product name (or 'done'): ")
    if name == "done":
        break
    units = int(input("Enter units sold: "))
    price = float(input("Enter unit price: "))
    product = Product(name, units, price)
    report.add_product(product)
report.print_report()
report.save_report()
```
